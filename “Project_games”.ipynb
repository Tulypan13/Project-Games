```python
import random
 
def show_rules2() -> None:
    """
    Shows the rules of the games using a menu of and the ability to move inside the menu.
    """
    # Welcome the user to the Rules of Games room.
    print("Welcome to the Rules of Games room!")
    # Show the available games and their corresponding numbers.
    print("Please choose one of the following games rules:")
    print("1. Rock, Paper, Scissors.")
    print("2. Rock, Paper, Scissors, Lizard, Spock.")
    print("3. Guess the number!")
    print("4. Exit.")
    
    # Keep asking the user for an option until they enter a valid one.
    while True:
        option = input("Enter the number of the game you want to check the rules: ")
        if option == '':
            print("Invalid input. Please enter a number from 1 to 4.")
        elif option.isdigit():
            option = int(option)
            if option < 1 or option > 4:
                print("Invalid input. Please enter a number from 1 to 4.")
            else:
                break
        else:
            print("Invalid input. Please enter a number from 1 to 4.")

    # Show the rules of the selected game and prompt the user to go back to the main menu.
    if option == 1:
        print("You have chosen: Rock, Paper, Scissors. The rules are:")
        print("The rules of the game are as follows:")
        print("- Rock crushes scissors.")
        print("- Paper covers rock.")
        print("- Scissors cuts paper.")        

        show_rules2()
    elif option == 2:
        print("You have chosen: Rock, Paper, Scissors, Lizard, Spock. The rules are:")
        print("The rules of the game are as follows:")
        print("- Rock crushes scissors and lizard.")
        print("- Paper covers rock and disproves spock.")
        print("- Scissors cuts paper and decapitates lizard.")
        print("- Lizard poisons spock and eats paper.")
        print("- Spock smashes scissors and vaporizes rock.")        

        show_rules2()
    elif option == 3:
        print("You have chosen: Guess the number! The rules are:")
        print("You have to guess a number from 1 to 10. The program will encourage you to take only 5 tries to guess.")
        print("You can still guess after 5 tries.")

        show_rules2()
    elif option == 4:
        print("Until next time! ")
       
    else:
        print("Invalid option. Please select a number from 1 to 4.")
    
def determinar_ganador2(opcion_usuario: str, opcion_programa: str) -> str: 
    """
    Determines the winner of a round of the rock, paper, scissors, lizard, spock game.

    Args:
        opcion_usuario (str): The option chosen by the user.
        opcion_programa (str): The option chosen by the program.

    Returns:
        str: The winner of the round: "user", "program" or "tie".
    """
    
    # Define the game rules as a dictionary where each key is a choice and its value is a list of choices that it beats.
    rules = {
        "rock": ["scissors", "lizard"],
        "paper": ["rock", "spock"],
        "scissors": ["paper", "lizard"],
        "lizard": ["paper", "spock"],
        "spock": ["rock", "scissors"]
    }
    
    # Check if the choices are the same, in which case the result is a tie.
    if opcion_usuario == opcion_programa:
        return "tie"
    
    # Check if the program's choice is in the list of choices that the user's choice beats, in which case the user wins.
    elif opcion_programa in rules[opcion_usuario]:
        return "user"
    
    # If the previous conditions are not met, the program wins.
    else:
        return "program"
    

def game2() -> None:
    """
    Main function of the rock, paper, scissors, lizard, spock game.
    Allows the user to play multiple rounds against the program.
    """
    options = ["rock", "paper", "scissors", "lizard", "spock"]  # create a list of options
    user_wins = 0  # initialize the user's win count to 0
    program_wins = 0  # initialize the program's win count to 0
    user_name = input("Enter your name: ")  # ask the user for their name
    num_rounds = int(input("Enter the number of rounds you want to play: "))  # ask the user for the number of rounds they want to play

    for i in range(num_rounds):  # loop through each round
        print(f"\nRound {i+1}")  # print the round number
        user_choice = input("Enter your choice (rock, paper, scissors, lizard, or spock): ").lower()  # ask the user for their choice
        while user_choice not in options:  # validate the user's choice
            user_choice = input("Invalid option. Enter your choice (rock, paper, scissors, lizard, or spock): ").lower()

        program_choice = random.choice(options)  # randomly choose the program's choice

        print(f"You chose {user_choice}, the program chose {program_choice}")  # print the choices
        round_winner = determinar_ganador2(user_choice, program_choice)  # determine the winner of the round

        if round_winner == "tie":  # handle a tie
            print("\n\t Tie\n")
        elif round_winner == "user":  # handle the user winning
            print("\n\t You win\n")
            user_wins += 1
        else:  # handle the program winning
            print("\n\t You lose\n")
            program_wins += 1

        print(f"{user_name}'s wins: {user_wins}")  # print the user's win count
        print(f"Program's wins: {program_wins}")  # print the program's win count

        if i < num_rounds - 1:  # ask the user if they want to play another round
            play_again = input("Press any key to play the next round, or 'q' to quit: ")
            if play_again.lower() == "q":
                break

    print("End of the game")  # print the end of the game message
         
def determinar_ganador1(user_option: str, program_option: str) -> str:
    """
    Determines the winner of a round of rock, paper, or scissors.

    Args:
        user_option (str): The option chosen by the user.
        program_option (str): The option chosen by the program.

    Returns:
        str: The winner of the round: "user", "program", or "tie".
    """
    # Rules of the game
    rules = {
        "rock": ["scissors"],
        "paper": ["rock"],
        "scissors": ["paper"]
    }
    # Check if the user_option is equal to the program_option, return "tie"
    if user_option == program_option:
        return "tie"
    # Check if the program_option is in the rules of the user_option, return "user"
    elif program_option in rules[user_option]:
        return "user"
    # Otherwise, return "program"
    else:
        return "program"
    
def game1() -> None:
    """
    Main function of the rock, paper, or scissors game.
    Allows the user to play multiple rounds against the program.
    """
    # Possible options
    options = ["rock", "paper", "scissors"]
    # Keep track of user and program wins
    user_wins = 0
    program_wins = 0
    # Ask for user's name and print the rules of the game
    user_name = input("Enter your name: ")
    print("This is a 3 round game")
    while user_name == "":
        user_name = input("Enter a valid name: ")
    
    round_count = 0
    # Play 3 rounds
    while round_count < 3:
        # Ask for user's option
        user_option = input("Enter your option (rock, paper, or scissors): ").lower()
        # Keep asking for user's option until a valid option is given
        while user_option not in options:
            user_option = input("Invalid option. Enter your option (rock, paper, or scissors): ").lower()
        
        # Generate a random option for the program
        program_option = random.choice(options)
        
        print(f"{user_name}, you chose {user_option}, the program chose {program_option}")
        # Determine the winner of the round and update the user and program win counts
        round_winner = determinar_ganador1(user_option, program_option)
        if round_winner == "tie":
            print("\n\t Tie\n")
        elif round_winner == "user":
            print("\n\t You won\n")
            user_wins += 1
        else:
            print("\n\t You lost\n")
            program_wins += 1
        
        # Print the current score
        print("Wins for", user_name + ":", user_wins)
        print("Wins for the program:", program_wins)
        
        # Ask the user if they want to play another round
        round_count += 1
        if round_count < 3:
            play_again = input("Press any key to play the next round, or 'q' to quit: ")
            if play_again.lower() == "q":
                break        
        
def game3():
    """
    This function allows the player to guess a random number between 1 and 10.
    The player has 5 attempts to guess the number correctly.
    If the player fails to guess the number in 5 attempts, they lose the game.
    The player can choose to play again or quit after each game.
    """
    
    while True: # loop to allow playing multiple times
        
        # generate a random number for the player to guess
        secret_number = random.randint(1, 10)
        guess = None
        num_tries = 0
        
        # loop to ask for the player's guess and give feedback on whether it's too high or too low
        while guess != secret_number:
            guess = int(input("Guess a number from 1 to 10: "))
            num_tries += 1
            
            if guess < secret_number:
                print("\n\t FAIL\n")
                print("Too low, try again!")
              
            elif guess > secret_number:
                print("\n\t FAIL\n")
                print("Too high, try again!")
           
            else:
                print(f"Congratulations! {guess} is correct!")
                print(f"You took {num_tries} tries.")
        
            # check if the player has used all 5 attempts
            if num_tries == 5:
                print("\nWARNING: You have used 5 attempts.")
                print("If you fail to guess the number on the next try, you lose the game.")
        
        # if the player used all 5 attempts, give them a warning
        if num_tries >= 5:
            print("\nYou've had 5 attempts or more. Think more carefully!")
        
        # ask if the player wants to play again, and exit the loop if they don't
        jugar_nuevamente = input("¿Want to play again? (y/n): ")
        if jugar_nuevamente.lower() != "y":
            break  
                
    print("End of the game")
    
    
def Play():
    """
    Main function of the project.
    Allows the user to play and select the game to play against the program.
    """
    # Welcomes the user to the game room
    print("Welcome to the game room!")

    # Displays a menu of available games
    print("Please choose one of the following games:")
    print("1. Rock, Paper, Scissors.")
    print("2. Rock, Paper, Scissors, Lizard, Spock.")
    print("3. Guess the number!")
    print("4. Rules of the games")
    print("5. Exit.")

    # Asks the user to select a game
    while True:
        option = input("Enter the number of the game you want to play: ")
        if option == '':
            print("Invalid input. Please enter a number from 1 to 5.")
        elif option.isdigit():
            option = int(option)
            if option < 1 or option > 5:
                print("Invalid input. Please enter a number from 1 to 5.")
            else:
                break
        else:
            print("Invalid input. Please enter a number from 1 to 5.")

    # Calls the corresponding function according to the selected option
    if option == 1:
        print("You have chosen: Rock, Paper, Scissors. Are you ready to test yourself?")
        game1()
        print("Thanks for playing! I hope you had a good time!")
        input("Press enter to go back to the main menu.")
        Play()
    elif option == 2:
        print("You have chosen: Rock, Paper, Scissors, Lizard, Spock. Are you ready to test yourself?")
        game2()
        print("Thanks for playing! I hope you had a good time!")
        input("Press enter to go back to the main menu.")
        Play()
    elif option == 3:
        print("You have chosen: Guess the number! Are you ready to test yourself?")
        game3()
        print("Thanks for playing! I hope you had a good time!")
        input("Press enter to go back to the main menu.")
        Play()
    elif option == 4:
        show_rules2()
        input("Press enter to go back to the main menu.")
        Play()
    elif option == 5:
        print("Until next time! ")
    else:
        print("Invalid option. Please select a number from 1 to 5.")
```
